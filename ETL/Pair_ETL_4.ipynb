```python
def crear_bbdd(nombre_bbdd):
    mydb = mysql.connector.connect(
                                    host= "localhost", 
                                    user= "root",
                                    password= "AlumnaAdalab")

    print("Conexión realizada con éxito.")

    mycursor = mydb.cursor()
    try:
        mycursor.execute(f"CREATE DATABASE IF NOT EXISTS {nombre_bbdd};")
        print(mycursor)

    except mysql.connector.Error as err:

        print(err)
        print("error code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)

 
```

```python
def crear_insertar_tabla(nombre_bbdd, query):

    cnx = mysql.connector.connect(user= "root", password= f"AlumnaAdalab", host= "127.0.0.1", database = f"{nombre_bbdd}" )

    mycursor = cnx.cursor()

    try:
        mycursor.execute(query)
        cnx.commit()

    except mysql.connector.Error as err:

        print(err)
        print("error code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)
```

Teníamos en mente que case_number fuera nuestra Primary Key, pero como tiene un nulo, aunque nuestro primer instinto fue eliminar el nulo, creemos que es mala práctica y vamos a crear otra Primary Key.

```python
query_tabla_ataques = """
CREATE TABLE IF NOT EXISTS `tiburones`.`ataques` (
  `index_ataques` INT NOT NULL AUTO_INCREMENT,
  `case_number` VARCHAR(45) NULL,
  `year` FLOAT NULL,
  `type` VARCHAR(200) NULL,
  `country` VARCHAR(45) NULL,
  `area` VARCHAR(200) NULL,
  `location` VARCHAR(200) NULL,
  `activity` VARCHAR(2000) NULL,
  `name` VARCHAR(200) NULL,
  `age` VARCHAR(45) NULL,
  `time` VARCHAR(20) NULL,
  `species_` VARCHAR(45) NULL,
  `injury` VARCHAR(2000) NULL,
  `fecha_limpia` VARCHAR(3) NULL,
  `fatal` VARCHAR(10) NULL,
  `sex` VARCHAR(10) NULL,
  `pais` VARCHAR(45) NULL,

  PRIMARY KEY (`index_ataques`));

"""
```

```python
crear_tabla_ataques = crear_insertar_tabla("tiburones", query_tabla_ataques)
```
